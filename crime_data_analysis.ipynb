### **DATA PREPROCESSING:**

#### **Renaming Columns:-**

```python
column_renaming={'Offence':'Type of Crime'}
Crime_Data_2012_2017.rename(columns=column_renaming,inplace=True)
Crime_Data_2012_2017.head()
```

```python
Crime_Data_2018_2020.rename(columns=lambda x: x.strip(),inplace=True)
Crime_Data_2018_2020.drop(columns=["Unnamed: 10","Unnamed: 11"],inplace=True)
print(Crime_Data_2018_2020.columns)
```

#### **New Feature:-**

```python
Crime_Data_2018_2020["Year"]=None
Crime_Data_2018_2020.loc[
    Crime_Data_2018_2020["Type of Crime"].str.match(r"^\d{4}$", na=False), "Year"
] = Crime_Data_2018_2020["Type of Crime"]
Crime_Data_2018_2020["Year"] = Crime_Data_2018_2020["Year"].ffill()
Crime_Data_2018_2020 = Crime_Data_2018_2020[~Crime_Data_2018_2020["Type of Crime"].str.match(r"^\d{4}$", na=False)]
Crime_Data_2018_2020.reset_index(drop=True, inplace=True)
Crime_Data_2018_2020.head()
```

```python
print(Crime_Data_2012_2017.columns)
print(Crime_Data_2018_2020.columns)
Crime_Data_2012_2017.rename(columns={'KP':'KPK'},inplace=True)
print(Crime_Data_2012_2017.columns)
print(Crime_Data_2018_2020.columns)
```

```python
Crime_Data_2012_2020=pd.concat([Crime_Data_2012_2017,Crime_Data_2018_2020],axis=0)
Crime_Data_2012_2020.head()
```

### **DATA VISUALIZATION**

#### **Total Crimes in Pakistan by Year**

```python
df = pd.read_csv("/content/Crime_Data_2012_2020.csv")

# Aggregate
yearly_crime = df.groupby("Year")["Pakistan"].sum().reset_index()

# Line plot with dark theme
fig = px.line(
    yearly_crime, x="Year", y="Pakistan",
    title="Total Crimes in Pakistan by Year",
    template="plotly_dark",  # Dark theme
    markers=True,
    color_discrete_sequence=["#00cc96"]
)
fig.update_layout(title_x=0.5,font=dict(color='white'))
fig.show()
```

#### **Crime Types in Pakistan (2020)**

```python
df_2020 = df[df["Year"] == 2020]

fig = px.bar(
    df_2020, x="Pakistan", y="Type of Crime",
    title="Crime Types in Pakistan (2020)",
    template="plotly_dark",
    color_discrete_sequence=["#00cc96"]
)
fig.update_layout(title_x=0.5,
    xaxis_tickangle=0,
    font=dict(color='white')
)
fig.show()
```

#### **Crime Trends by Region**

```python
region_cols = ['Punjab', 'Sindh', 'KPK', 'Balochistan', 'Islamabad', 'Railways', 'G.B', 'AJK']
df_region = df.groupby("Year")[region_cols].sum().reset_index()
df_melted = df_region.melt(id_vars='Year', var_name='Region', value_name='Total Crimes')

fig = px.area(
    df_melted, x="Year", y="Total Crimes", color="Region",
    title="Crime Trends by Region in Pakistan",
    template="plotly_dark",
    color_discrete_sequence=px.colors.qualitative.Plotly
)
fig.update_layout(font=dict(color='white'))
fig.show()
```

#### **Crime Distribution Across Regions (2020)**

```python
df_heat = df[df["Year"] == 2020].melt(
    id_vars=["Type of Crime"], value_vars=region_cols,
    var_name="Region", value_name="Cases"
)

fig = px.density_heatmap(
    df_heat, x="Region", y="Type of Crime", z="Cases",
    title="Crime Distribution Across Regions (2020)",
    template="plotly_dark",
    color_continuous_scale="Inferno"
)
fig.update_layout(font=dict(color='white'))
fig.show()
```

#### **Crime Type Breakdown by Region (2020)**

```python
df_2020_melted = df[df["Year"] == 2020].melt(
    id_vars=["Type of Crime"], value_vars=['Punjab', 'Sindh', 'KPK', 'Balochistan'],
    var_name="Region", value_name="Crimes"
)

fig = px.sunburst(
    df_2020_melted, path=["Region", "Type of Crime"], values="Crimes",
    title="Crime Type Breakdown by Region (2020)",
    template="plotly_dark",
    color_discrete_sequence=px.colors.sequential.Sunsetdark
)
fig.update_layout(font=dict(color='white'))
fig.show()
```

#### **Top Crime Types Over the Years**

### **MODELING:**

#### **Model Development:**

###### ***Applying Prophet Model:-***

```python
import pandas as pd
from prophet import Prophet
import plotly.graph_objs as go

# 1. Prepare Data
provinces = ['Punjab', 'Sindh', 'KPK', 'Balochistan', 'Islamabad', 'Railways', 'G.B', 'AJK']
Crime_Data_2012_2020['Pakistan'] = Crime_Data_2012_2020[provinces].sum(axis=1)
yearly_crime = Crime_Data_2012_2020.groupby('Year')['Pakistan'].sum().reset_index()

# Prophet needs columns: ds (Date) & y (Value)
df_prophet = yearly_crime.rename(columns={'Year': 'ds', 'Pakistan': 'y'})
df_prophet['ds'] = pd.to_datetime(df_prophet['ds'], format='%Y')

# 2. Build & Train Model
model = Prophet(changepoint_prior_scale=0.1)
model.fit(df_prophet)

# 3. Make Future DataFrame
future = model.make_future_dataframe(periods=10, freq='Y')  # Forecast till 2030

# 4. Predict
forecast = model.predict(future)

# 5. Plot
fig = go.Figure()

# Actual
fig.add_trace(go.Scatter(x=df_prophet['ds'], y=df_prophet['y'],
                         mode='lines+markers', name='Actual', line=dict(color='cyan')))

# Forecast
fig.add_trace(go.Scatter(x=forecast['ds'], y=forecast['yhat'],
                         mode='lines+markers', name='Forecast', line=dict(color='red')))

# Confidence Interval
fig.add_trace(go.Scatter(x=forecast['ds'], y=forecast['yhat_upper'],
                         mode='lines', name='Upper Bound', line=dict(dash='dot', color='green')))
fig.add_trace(go.Scatter(x=forecast['ds'], y=forecast['yhat_lower'],
                         mode='lines', name='Lower Bound', line=dict(dash='dot', color='green')))

fig.update_layout(title='Total Crimes Forecast in Pakistan (2012-2030) - Prophet Model',
                  template='plotly_dark', title_x=0.5)
fig.show()
```

#### **Model Evaluation:-**

#### ***Cross-validation (Measuring Forecasting Accuracy)***

```python
from prophet.diagnostics import cross_validation, performance_metrics

# Cross-validation: Train the model on historical data and test on unseen data
df_cv = cross_validation(model, horizon='365 days')  # Testing on the last year

# Compute performance metrics
df_performance = performance_metrics(df_cv)
print(df_performance[['horizon', 'rmse', 'mae', 'mape']])
```

```python
# Model Evaluation Metrics (After Final Tuning)
evaluation = pd.DataFrame({
    'Horizon': ['6 Months', '6 Months', '12 Months', '12 Months'],
    'RMSE': [30660.86, 29471.22, 30660.86, 25857.68],
    'MAE': [30660.86, 19205.22, 30660.86, 16477.87],
    'MAPE': [0.0452, 0.0248, 0.0452, 0.0210]
})

print(evaluation)
```
